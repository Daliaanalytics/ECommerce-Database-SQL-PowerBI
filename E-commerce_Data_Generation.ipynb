```python
import pandas as pd
import numpy as np
from faker import Faker
import random
from datetime import datetime, timedelta

fake=Faker()
random.seed(42)
np.random.seed(42)
```

```python
categories=  pd.DataFrame({
    'Category_id': range(1,9),
    'Category_name': ['Electronics', 'Clothing', 'Home & Kitchen',
        'Sports', 'Books', 'Beauty', 'Toys', 'Furniture']
})
categories
```

```python
customers= pd.DataFrame({
    'customer_id': range(1,501),
    'first_name': [fake.first_name() for _ in range(500)],
    'last_name':[fake.last_name() for _ in range(500)],
    'email': [fake.email() for _ in range(500)],
    'city': [fake.city() for _ in range(500)],
    'country': [fake.country() for _ in range(500)],
    'registration_date': [fake.date_between(start_date= '-3y', end_date= 'today') for _ in range(500)]
        
})
customers.head()
```

```python
f"Customers no: {len(customers)}"
```

```python
products = pd.DataFrame({
    'product_id': range(1, 201),
    'product_name': [fake.catch_phrase() for _ in range(200)],
    'category_id': [random.randint(1, 8) for _ in range(200)],
    'price': [round(random.uniform(5, 500), 2) for _ in range(200)],
    'stock_quantity': [random.randint(0, 100) for _ in range(200)]
})

```

```python
products.head()
```

```python
f"Products no: {len(products)}"
```

```python
orders = pd.DataFrame({
    'order_id': range(1, 1001),
    'customer_id': [random.randint(1, 500) for _ in range(1000)],
    'order_date': [
        fake.date_between(start_date='-2y', end_date='today')
        for _ in range(1000)
    ],
    'status': [
        random.choice(['Completed', 'Pending', 'Cancelled', 'Refunded'])
        for _ in range(1000)
    ],
    'total_amount': [round(random.uniform(10, 2000), 2) for _ in range(1000)]
})
```

```python
orders.head()
```

```python
f"Orders no: {len(orders)}"
```

```python
order_items = []

for order_id in range(1, 1001):
    num_items = random.randint(1, 5)
    products_chosen = random.sample(range(1, 201), num_items)
    
    for product_id in products_chosen:
        quantity = random.randint(1, 10)
        unit_price = round(random.uniform(5, 500), 2)
        order_items.append({
            'order_item_id': len(order_items) + 1,
            'order_id': order_id,
            'product_id': product_id,
            'quantity': quantity,
            'unit_price': unit_price
        })

order_items = pd.DataFrame(order_items)

```
